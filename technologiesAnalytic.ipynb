```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('technologies.csv')
skills_list = df['skills'].apply(lambda x: x.split(',')).explode()

skills_count = skills_list.value_counts().reset_index()
skills_count.columns = ['Skill', 'Count']

skills_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(skills_count['Skill'], skills_count['Count'], color='skyblue')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.title('Skill Counts')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
